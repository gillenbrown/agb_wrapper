```python
hatches = ["\\", "/", "O", "."]
labels = ["DE", "CE", "C", "D"]
```

```python
fig, ax = bpl.subplots()

for tab, hatch, label in zip(tabs, hatches, labels):
    values = [tab.random_binomial_py(0.5, 20) for _ in range(10000000)]
    
    ax.hist(values, bin_size=1.0, histtype="step", hatch=hatch, label=label)
    
    print(max(values), min(values))

ax.legend()
```
